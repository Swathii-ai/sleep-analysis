```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/user/Downloads/Sleep_Efficiency.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
#Awakenings,Caffeine consumption,Alcohol consumption,Exercise frequency
```

```python
a = data["Awakenings"].mean()
c_c = data["Caffeine consumption"].mean()
a_c = data["Alcohol consumption"].mean()
e_f = data["Exercise frequency"].mean()
```

```python
data.fillna({"Awakenings":a,"Caffeine consumption":c_c,"Alcohol consumption":a_c,"Exercise frequency":e_f},inplace = True)
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
col =["Gender","Bedtime","Wakeup time","Smoking status"]
```

```python
for i in col:
    data[i] = le.fit_transform(data[i])
data
```

```python
x = data.drop("Sleep efficiency",axis=1)
y = data["Sleep efficiency"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size = 0.8)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
linear_pred = lr.predict(xtest)
linear_pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(linear_pred,ytest,color = "brown")
plt.plot([min(ytest),max(ytest)],[min(ytest),max(ytest)],color = "yellow")
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_se = mean_squared_error(linear_pred,ytest)
mean_ae = mean_absolute_error(linear_pred,ytest)
r2_s = r2_score(linear_pred,ytest)
```

```python
print("Mean Squared Error:",mean_se)
print("Mean Absolute Error:",mean_ae)
print("r2 Score:",r2_s)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(xtrain,ytrain)
```

```python
dt_p = dt.predict(xtest)
dt_p
```

```python
plt.scatter(dt_p,ytest)
plt.plot([min(ytest),max(ytest)],[min(ytest),max(ytest)],color = "brown")
```

```python
s1 = mean_squared_error(dt_p,ytest)
a2 = mean_absolute_error(dt_p,ytest)
r1 = r2_score(dt_p,ytest)
```

```python
print("Mean Squared Error:",s1)
print("Mean Absolute Error:",a2)
print("r2 Score:",r1)
```

```python
from sklearn.ensemble import RandomForestRegressor
r_f = RandomForestRegressor()
r_f.fit(xtrain,ytrain)
```

```python
rf_pred = r_f.predict(xtest)
plt.scatter(rf_pred,ytest)
plt.plot([min(ytest),max(ytest)],[min(ytest),max(ytest)])           
```

```python
s2 = mean_squared_error(rf_pred,ytest)
a2 = mean_absolute_error(rf_pred,ytest)
r2 = r2_score(rf_pred,ytest)
print("Mean Squared Error:",s2)
print("Mean absolute Error:",a2)
print("r2 score:",r2)
```
